### Linear Regression

```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
```

### Min-Max Normalization

### X' = (X - X_Min)/(X_Max - X_Min)

```python
#Preprocessing input data
#Read data from csv. Dataset.csv contains  rows from (n−1) ∗ 20 + 2 to n ∗ 20 + 1 from file 
#Data Take Home Assignment 1 Exercise A. n = 5 since Group 5
data = pd.read_csv('Dataset.csv')
x= data.iloc[:,0]
y = data.iloc[:,1]

#Applying Min-Max normalization on data
#Data transformation - Min- Max Normalization
x = ((x - x.min())/ (x.max() - x.min()))
y = ((y - y.min())/ (y.max() - y.min()))

#plotting transformed data
plt.scatter(x,y)
plt.xlabel("Temperature(Celcius)")
plt.ylabel("Net hourly electrical energy output(MW)")
plt.show()
```

### Least Squares
#### Linear equation for least squares can be given as y = mx + b
#### m = ((x - x_mean)*(y-y_mean))/((x-x_mean)^2)
#### b = (y_mean - m*x_mean)

```python
#Building Linear regression with least squares
x_mean = np.mean(x)
y_mean = np.mean(y)

num = 0
den = 0

for i in range(len(x)):
    num += (x[i]-x_mean)*(y[i]-y_mean)
    den += (x[i]-x_mean)**2

m = num /den
b= y_mean - m*x_mean

print(f'm derived from least square method is {m}')
print(f'b derived from least square method is {b}')
```

```python
#Making Predictions
y_pred = m*x + b
plt.scatter(x,y) #actual
plt.scatter(x, y_pred, color = 'red') #predicted
plt.plot(x,y_pred, color = 'red')
plt.xlabel("Temperature(Celcius)")
plt.ylabel("Net hourly electrical energy output(MW)")
plt.show()
```

### Cost Function 
### Mean Squared Error = (1/n)*sum ((yi-(mxi+b))^2)
### d/dm = (2/n)*sum(-xi(yi-(mxi+b)))
### d/db = (2/n)*sum(-(yi-mxi+b))
### m = m- learning rate * (d/dm)
### b = b - learning rate * (d/db)

```python
#Gradient descent initial values
x = np.array(x)
y = np.array(y)
m_init = b_init = 0
n = len(x)
learning_rate = 0.01

#First iteration
y_pred = m_init*x + b_init
cost = (1/n)* sum([val**2 for val in (y-y_pred)])
dm = -(2/n)*sum(x*(y-y_pred))
db = -(2/n)*sum(y-y_pred)
m_init = m_init - learning_rate * dm
b_init = b_init - learning_rate * db
print("After first iteration m {}, b {}, cost {}".format(m_init, b_init, cost))
    
#Making Predictions
y_pred = m_init*x + b_init

#After first iteration plotting the regression line
print("After first iteration graph of regression line looks like: ")
plt.xlabel("Temperature(Celcius)")
plt.ylabel("Net hourly electrical energy output(MW)")
plt.scatter(x,y)
plt.scatter(x, y_pred)
plt.plot(x,y_pred, color = 'red')
plt.show()
```

```python
#Second iteration
y_pred = m_init*x + b_init
cost = (1/n)* sum([val**2 for val in (y-y_pred)])
dm = -(2/n)*sum(x*(y-y_pred))
db = -(2/n)*sum(y-y_pred)
m_init = m_init - learning_rate * dm
b_init = b_init - learning_rate * db
print("After second iteration m {}, b {}, cost {}".format(m_init, b_init, cost))
    
#Making Predictions
y_pred = m_init*x + b_init

#After second iteration plotting the regression line
print("After second iteration graph of regression line looks like: ")
plt.xlabel("Temperature(Celcius)")
plt.ylabel("Net hourly electrical energy output(MW)")
plt.scatter(x,y)
plt.scatter(x, y_pred)
plt.plot(x,y_pred, color = 'red')
plt.show()
```

```python
#Third iteration
y_pred = m_init*x + b_init
cost = (1/n)* sum([val**2 for val in (y-y_pred)])
dm = -(2/n)*sum(x*(y-y_pred))
db = -(2/n)*sum(y-y_pred)
m_init = m_init - learning_rate * dm
b_init = b_init - learning_rate * db
print("After third iteration m {}, b {}, cost {}".format(m_init, b_init, cost))
    
#Making Predictions
y_pred = m_init*x + b_init

#After third iteration plotting the regression line
print("After third iteration graph of regression line looks like: ")
plt.xlabel("Temperature(Celcius)")
plt.ylabel("Net hourly electrical energy output(MW)")
plt.scatter(x,y)
plt.scatter(x, y_pred)
plt.plot(x,y_pred, color = 'red')
plt.show()
```

```python
#Gradient descent
iterations = 12220
cost_list = []

for i in range(iterations):
    y_pred = m_init*x + b_init
    cost = (1/n)* sum([val**2 for val in (y-y_pred)])
    cost_list.append(cost)
    dm = -(2/n)*sum(x*(y-y_pred))
    db = -(2/n)*sum(y-y_pred)
    m_init = m_init - learning_rate * dm
    b_init = b_init - learning_rate * db
    print("m {}, b {}, cost {}, iterations {}".format(m_init, b_init, cost, i))

print('Cost at each iteration:')
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.plot(list(range(iterations)), cost_list, '-r') #plot the cost function.
```

```python
#Making Predictions
y_pred = m_init*x + b_init

#plotting graph after last iteration
print("After last iteration regression line looks like: ")
plt.xlabel("Temperature(Celcius)")
plt.ylabel("Net hourly electrical energy output(MW)")
plt.scatter(x,y)
plt.scatter(x, y_pred)
plt.plot(x,y_pred, color = 'red')
plt.show()
```
